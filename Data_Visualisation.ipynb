**Line Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
y1 = x**2
y2 = 2*x+3

plt.figure(figsize=(7,7))
plt.plot(x,y1,color='red',label = 'apple', linestyle = "dashed")
plt.plot(x,y2, color='green',label = 'mango', marker = "o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

**Scatter Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
y1 = x**2
y2 = 2*x+3

plt.figure(figsize=(7,7)) #adjusting the size of the plot 

plt.scatter(x,y1,color='red',label = 'apple', marker = "x")
plt.scatter(x,y2, color='green',label = 'mango', marker = "o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Prices of fruits over Time")
plt.legend()
plt.show()
```

**Bar Graph**

```python
x_coordinates = np.array([0,1,2])

plt.figure(figsize=(7,7))
plt.bar(x_coordinates, [10,20,15], width = 0.25, label = "Current Year", tick_label = ["Gold", "Silver", "Platinum"])
plt.bar(x_coordinates+0.25, [5,10,20],width = 0.25, label = "Next Year")
plt.xlabel("Metals in mg")
plt.ylabel("Price in 10 lakhs")
plt.title("Bar Graph of Metal Prices")
plt.ylim(0,30)
plt.legend()
plt.show()
```

# **Movie Data Visualisation**

We are given a movie dataset. We have to plot on the X axis the length of the title of the movie vs the Y axis which tells the count of the number of movies having the particular length given on Y axis.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/2.%20Working%20with%20Libraries/movie_metadata.csv'
df = pd.read_csv(url)
```

```python
df.head(50)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
print(titles)
```

```python
freq_titles = {}

for title in titles:
  length = len(title)
  
  if freq_titles.get(length) is None:
    freq_titles[length] = 1
  
  else: 
    freq_titles[length] += 1

print(freq_titles) #prints {length of the movies : number of movies having that length}
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))

plt.scatter(X,Y)
plt.xlabel("Length of Movie Characters")
plt.ylabel("Number of Movies")
plt.title("Movie Data Visualisation")
plt.show()
```
